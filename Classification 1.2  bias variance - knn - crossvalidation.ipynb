```
# Load Iris and organize our data
from sklearn.datasets import load_iris
data = load_iris()
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']
```

```
# Î‘Ï‚ Ï„Ï…Ï€ÏÏƒÎ¿Ï…Î¼Îµ Ï„Î± Î¿Î½ÏŒÎ¼Î±Ï„Î± Ï„Ï‰Î½ Ï‡Î±ÏÎºÏ„Î·ÏÎ¹ÏƒÏ„Î¹ÎºÏÎ½
print(feature_names)
print(label_names)
```

```
# Î§ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Ï„Î· Î³Î½Ï‰ÏƒÏ„Î® train_test_split Î³Î¹Î± Î½Î± Î´Î¹Î±Ï‡Ï‰ÏÎ¯ÏƒÎ¿Ï…Î¼Îµ ÏƒÎµ train ÎºÎ±Î¹ test set
# Ï„Î¿ (int) ÏŒÏÎ¹ÏƒÎ¼Î± "random_state" ÎµÎ¯Î½Î±Î¹ Ï„Î¿ seed Ï„Î·Ï‚ Î³ÎµÎ½Î½Î®Ï„ÏÎ¹Î±Ï‚ Ï„Ï…Ï‡Î±Î¯Ï‰Î½ Î±ÏÎ¹Î¸Î¼ÏÎ½
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.40, random_state=78)
```

Î”Î¹Î±Î»Î­Î³Î¿Ï…Î¼Îµ Ï„Ï…Ï‡Î±Î¯Î± Ï„Î¹Î¼Î® 5 Î³Î¹Î± Ï„Î·Î½ Ï…Ï€ÎµÏÏ€Î±ÏÎ¬Î¼ÎµÏ„ÏÎ¿ k, ÎµÎºÏ€Î±Î¹Î´ÎµÏÎ¿Ï…Î¼Îµ Î­Î½Î± kNN classifier ÎºÎ±Î¹ Ï…Ï€Î¿Î»Î¿Î³Î¯Î¶Î¿Ï…Î¼Îµ Ï„Î·Î½ Ï€Î¹ÏƒÏ„ÏŒÏ„Î·Ï„Î± Ï„Î¿Ï… ÏƒÏ„Î¿ Iris, ÏƒÏ„Î¿ Ï€ÏÎ¿Î·Î³Î¿ÏÎ¼ÎµÎ½Î¿ train/test split

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print(accuracy_score(y_test, pred))
```

Î“Î¹Î± Î½Î± Î±ÎºÎ¿Î»Î¿Ï…Î¸Î®ÏƒÎ¿Ï…Î¼Îµ ÏƒÏ‰ÏƒÏ„Î¬ Ï„Î¿ Ï€ÏÏ‰Ï„ÏŒÎºÎ¿Î»Î»Î¿, Î±Ï…Ï„ÏŒ Ï€Î¿Ï… Ï€ÏÎ­Ï€ÎµÎ¹ Î½Î± ÎºÎ¬Î½Î¿Ï…Î¼Îµ ÎµÎ¯Î½Î±Î¹ Î½Î± Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î®ÏƒÎ¿Ï…Î¼Îµ Î¼ÏŒÎ½Î¿ Ï„Î¿ Ï€ÏÎ±Î³Î¼Î±Ï„Î¹ÎºÏŒ training set Î³Î¹Î± Î½Î± Î´Î¹Î±Î»Î­Î¾Î¿Ï…Î¼Îµ Ï„Î¹Ï‚ Î²Î­Î»Ï„Î¹ÏƒÏ„ÎµÏ‚ Ï…Ï€ÎµÏÏ€Î±ÏÎ±Î¼Î­Ï„ÏÎ¿Ï…Ï‚. Î˜Î± Î¼Ï€Î¿ÏÎ¿ÏÏƒÎ±Î¼Îµ Î½Î± ÎºÏÎ±Ï„Î®ÏƒÎ¿Ï…Î¼Îµ Î­Î½Î± Ï€Î¿ÏƒÎ¿ÏƒÏ„ÏŒ Î´ÎµÎ¹Î³Î¼Î¬Ï„Ï‰Î½ Ï‰Ï‚ ÏƒÏÎ½Î¿Î»Î¿ ÎµÏ€Î¹ÎºÏÏÏ‰ÏƒÎ·Ï‚ (validation set Ï€Ï‡ Î¬Î»Î»Î¿ Î­Î½Î± 1/3) Ï„Î¿Ï… training set ÎºÎ±Î¹ Î½Î± Î±ÎºÎ¿Î»Î¿Ï…Î¸Î®ÏƒÎ¿Ï…Î¼Îµ Ï„Î·Î½ Ï€ÏÎ¿Î·Î³Î¿ÏÎ¼ÎµÎ½Î· Î´Î¹Î±Î´Î¹ÎºÎ±ÏƒÎ¯Î±: ÎµÎºÏ€Î±Î¯Î´ÎµÏ…ÏƒÎ· ÏƒÏ„Î¿ 1/3 training set, ÎµÏ€Î¹ÎºÏÏÏ‰ÏƒÎ· ÏƒÎµ 1/3 ÎºÎ±Î¹ Ï„ÎµÎ»Î¹ÎºÎ¬ Î±Î¾Î¹Î¿Î»ÏŒÎ³Î·ÏƒÎ· ÏƒÏ„Î¿ 1/3 data set. Î©ÏƒÏ„ÏŒÏƒÎ¿ Î±Ï…Ï„Î® Î· Î¼ÎµÎ¸Î¿Î´Î¿Î»Î¿Î³Î¯Î± "Î±Ï‡ÏÎ·ÏƒÏ„ÎµÏÎµÎ¹" Î¼ÎµÎ³Î¬Î»Î¿ Î¼Î­ÏÎ¿Ï‚ Ï„Î¿Ï… dataset (Ï„Î± 2/3) Ï‰Ï‚ Ï€ÏÎ¿Ï‚ Ï„Î·Î½ ÎµÎºÏ€Î±Î¯Î´ÎµÏ…ÏƒÎ· Ï„Î¿Ï… Ï„Î±Î¾Î¹Î½Î¿Î¼Î·Ï„Î®. Î ÏÎ±ÎºÏ„Î¹ÎºÎ¬ Î»Î¿Î¹Ï€ÏŒÎ½, Ï€ÏÎ¿Ï„Î¹Î¼Î¿ÏÎ¼Îµ Î½Î± Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Ï„Î· Î¼Î­Î¸Î¿Î´Î¿ Ï„Î·Ï‚ Î´Î¹Î±ÏƒÏ„Î±Ï…ÏÎ¿ÏÎ¼ÎµÎ½Î·Ï‚ ÎµÏ€Î¹ÎºÏÏÏ‰ÏƒÎ·Ï‚ (Cross Validation).

Î£Ï„Î¿ Cross Validation Î±ÏÏ‡Î¹ÎºÎ¬ Ï‡Ï‰ÏÎ¯Î¶Î¿Ï…Î¼Îµ Ï„Î¿ training set ÏƒÎµ Î­Î½Î±Î½ Î±ÏÎ¹Î¸Î¼ÏŒ "Ï€Ï„Ï…Ï‡ÏÎ½" (folds). Î£Ï…Î½Î·Î¸Î¹ÏƒÎ¼Î­Î½ÎµÏ‚ Ï„Î¹Î¼Î­Ï‚ ÎµÎ¯Î½Î±Î¹ Ï„Î¿ 5 ÎºÎ±Î¹ Ï„Î¿ 10 (5-fold ÎºÎ±Î¹ 10-fold CV). Î£Ï„Î· ÏƒÏ…Î½Î­Ï‡ÎµÎ¹Î±, Î³Î¹Î± ÎºÎ¬Î¸Îµ k-fold (Î¬ÏƒÏ‡ÎµÏ„Î¿ Î±Ï€ÏŒ Ï„Î¿ k Ï„Î¿Ï… kNN), Î¸ÎµÏ‰ÏÎ¿ÏÎ¼Îµ ÏŒÏ„Î¹ Ï„Î± k Î¼ÎµÎ¯Î¿Î½ 1 folds ÎµÎ¯Î½Î±Î¹ training set ÎºÎ±Î¹ ÏŒÏ„Î¹ Ï„Î¿ fold Ï€Î¿Ï… Î±Ï†Î®ÏƒÎ±Î¼Îµ Î­Î¾Ï‰ ÎµÎ¯Î½Î±Î¹ Ï„Î¿ test set. Î¥Ï€Î¿Î»Î¿Î³Î¯Î¶Î¿Ï…Î¼Îµ Ï„Î· Î¼ÎµÏ„ÏÎ¹ÎºÎ® ÏƒÏ†Î¬Î»Î¼Î±Ï„ÏŒÏ‚ Î¼Î±Ï‚ ÏƒÏ„Î¿ test set Ï€Î¿Ï… Î¿ÏÎ¯Î¶ÎµÎ¹ Ï„Î¿ fold. Î•Ï€Î±Î½Î±Î»Î±Î¼Î²Î¬Î½Î¿Ï…Î¼Îµ Ï„Î· Î´Î¹Î±Î´Î¹ÎºÎ±ÏƒÎ¯Î± Î³Î¹Î± Ï„Î± k folds Î³Î¹Î± ÎºÎ¬Î¸Îµ Ï„Î¹Î¼Î® Ï„Ï‰Î½ Ï…Ï€ÎµÏÏ€Î±ÏÎ±Î¼Î­Ï„ÏÏ‰Î½ ÎºÎ±Î¹ Ï…Ï€Î¿Î»Î¿Î³Î¯Î¶Î¿Ï…Î¼Îµ Ï„Î· Î¼Î­ÏƒÎ· Ï„Î¹Î¼Î® Ï„Î·Ï‚ Î¼ÎµÏ„ÏÎ¹ÎºÎ®Ï‚ Ï„Î¿Ï… ÏƒÏ†Î¬Î»Î¼Î±Ï„Î¿Ï‚. ÎœÎµ Î±Ï…Ï„ÏŒ Ï„Î¿Î½ Ï„ÏÏŒÏ€Î¿, Î±Ï†ÎµÎ½ÏŒÏ‚ ÎµÎ¯Î¼Î±ÏƒÏ„Îµ Î±Î¼ÎµÏÏŒÎ»Î·Ï€Ï„Î¿Î¹ ÏƒÏ„Î·Î½ Î±Î¾Î¹Î¿Î»ÏŒÎ³Î·ÏƒÎ· Î±Ï†Î®Î½Î¿Î½Ï„Î±Ï‚ Ï„ÎµÎ»ÎµÎ¯Ï‰Ï‚ Î­Î¾Ï‰ Ï„Î¿ test set ÎºÎ±Î¹ Î±Ï†ÎµÏ„Î­ÏÎ¿Ï… Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Î±Ï€Î¿Ï„ÎµÎ»ÎµÏƒÎ¼Î±Ï„Î¹ÎºÎ¬ Ï„Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î± ÎµÎºÏ€Î±Î¯Î´ÎµÏ…ÏƒÎ·Ï‚: Ï„Î± Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ ÏŒÎ»Î± ÎºÎ±Î¹ Ï€Î±Î¯ÏÎ½Î¿Î½Ï„Î±Ï‚ Ï„Î· Î¼Î­ÏƒÎ· Ï„Î¹Î¼Î® ÎµÎ¾Î±Î»ÎµÎ¯Ï†Î¿Ï…Î¼Îµ Ï€Î¹Î¸Î±Î½Î­Ï‚ Î±Î½Ï‰Î¼Î±Î»Î¯ÎµÏ‚ ÏƒÏ„Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î±.

![Cross validation](https://sebastianraschka.com/images/faq/evaluate-a-model/k-fold.png "Cross Validation")

To Scikit Learn Î­Ï‡ÎµÎ¹ ÏƒÏ…Î½Î±ÏÏ„Î®ÏƒÎµÎ¹Ï‚ Î³Î¹Î± Î½Î± ÎºÎ¬Î½ÎµÎ¹ Î±Ï…Ï„ÏŒÎ¼Î±Ï„Î± cross validation (Î½Î± Î¿ÏÎ¯Î¶ÎµÎ¹ folds ÎºÎ±Î¹ Î½Î±  Ï…Ï€Î¿Î»Î¿Î³Î¯Î¶ÎµÎ¹ Ï„Î¹Î¼Î­Ï‚ ÎºÎ±Î¹ Î¼Î­ÏƒÎ¿Ï…Ï‚ ÏŒÏÎ¿Ï…Ï‚). Î˜Î± ÎºÎ¬Î½Î¿Ï…Î¼Îµ 5 fold cross validation Î³Î¹Î± Î½Î± Ï…Ï€Î¿Î»Î¿Î³Î¯ÏƒÎ¿Ï…Î¼Îµ Ï„Î¿ Î²Î­Î»Ï„Î¹ÏƒÏ„Î¿ k Ï„Î¿Ï… kNN ÏƒÏ„Î¿ Iris.

```
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Ï†Ï„Î¹Î¬Ï‡Î½Î¿Ï…Î¼Îµ Î¼Î¹Î± Î»Î¯ÏƒÏ„Î± Î±Ï€ÏŒ Ï„Î¿ 1 Î­Ï‰Ï‚ Ï„Î¿ 50
myList = list(range(1,50))
# ÎšÏÎ±Ï„Î¬Î¼Îµ Î¼ÏŒÎ½Î¿ Ï„Î± Ï€ÎµÏÎ¹Ï„Ï„Î¬ k
neighbors = list(filter(lambda x: x % 2 != 0, myList))
# empty list that will hold cv scores
cv_scores = []
# perform 5-fold cross validation
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())
```

Î˜Î± Ï„Ï…Ï€ÏÏƒÎ¿Ï…Î¼Îµ Ï„Î·Î½ Ï„Î¹Î¼Î® Ï„Î¿Ï… ÏƒÏ†Î¬Î»Î¼Î±Ï„Î¿Ï‚ ÏƒÎµ Î³ÏÎ±Ï†Î¹ÎºÎ® Ï€Î±ÏÎ¬ÏƒÏ„Î±ÏƒÎ· Ï‰Ï‚ ÏƒÏ…Î½Î¬ÏÏ„Î·ÏƒÎ· Ï„Î¿Ï… k Î³Î¹Î± Ï„Î¿ training set, ÎºÎ±Î¹ Ï„Î·Î½ Ï„Î¹Î¼Î® Ï„Î·Ï‚ Ï€Î¹ÏƒÏ„ÏŒÏ„Î·Ï„Î±Ï‚ ÏƒÏ„Î¿ test set Î³Î¹Î± Ï„Î¿ Î²Î­Î»Ï„Î¹ÏƒÏ„Î¿ k (Ï…Ï€Î¿Î»Î¿Î³Î¹ÏƒÎ¼Î­Î½Î¿ ÏƒÏ„Î¿ training set)

```
# ÎšÎ¬Î½Î¿Ï…Î¼Îµ update Ï„Î·Î½ matplotlib, Î¯ÏƒÏ‰Ï‚ Ï‡ÏÎµÎ¹Î±ÏƒÏ„ÎµÎ¯ restart Ï„Î¿Ï… runtime
!pip install --upgrade matplotlib
```

```
# ÎšÎ¬Î½Î¿Ï…Î¼Îµ import Ï„Î·Î½ matplotplib
import matplotlib.pyplot as plt

# Ï„Î¿ ÏƒÏ†Î¬Î»Î¼Î± ÎµÎ¯Î½Î±Î¹ Ï„Î¿ Î±Î½Ï„Î¯ÏƒÏ„ÏÎ¿Ï†Î¿ Ï„Î·Ï‚ Ï€Î¹ÏƒÏ„ÏŒÏ„Î·Ï„Î±Ï‚
mean_error = [1 - x for x in cv_scores]

# plot misclassification error vs k
plt.plot(neighbors, mean_error)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()

# determining best k
optimal_k = neighbors[mean_error.index(min(mean_error))]
print("The optimal number of neighbors (calculated in the training set) is %d" % optimal_k)

# Î³Î¹Î± Ï„Î¿ optimal k Ï€Î±Î¯ÏÎ½Î¿Ï…Î¼Îµ ÎºÎ±Î¹ Ï„Î± Î±Ï€Î¿Ï„Î­Î»ÎµÏƒÎ¼Î±Ï„Î± ÏƒÏ„Î¿ test set
knn = KNeighborsClassifier(n_neighbors = optimal_k)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print("\nOptimal accuracy on the test set is", accuracy_score(y_test, pred), "with k=", optimal_k)
```
